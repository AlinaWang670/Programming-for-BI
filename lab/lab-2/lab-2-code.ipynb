## Exceptions
An exception is an event which occurs during the execution of a program and that disrupts the normal flow of the program's instructions. The default behaviour for exceptions is to terminate execution


````python
def check_index(v,i):
    return(v[i])

x = [0,1,2,3]

try: 
    check_index(x, 5) 
except IndexError: 
    print('got exception') 
````

Code that may raise an exception are typically stated in a ``try-except-else`` statement:
```python
try:
		statements
except name1:
		statements
except (name2, name3):
		statements
except name4 as var:
		statements
except:
		statements
else:
		statements
finally:
		statements
```

- except 
    * A single try statement can have multiple except statements. Useful when the `try` block contains statements that may throw different types of exceptions
    * The except statements can catch a single exception or a list of exceptions
    * The exception can be assigned to a variable
    * You can also provide a generic except clause, which handles any exception
- else
    * After the `except` clause(s), you can include an `else` clause which executes if the code in the try block does not raise an exception
- finally 
    * The `finally` block is always run, regardless if an exception occurs or not

#### Built-in Exceptions
Python has different types of exceptions.

````python
def dividing(x,y):
    if y == 0.0:
        raise ZeroDivisionError()
    else:
        return(x/y)

a = 3
b = 0
    
try:
    result = dividing(a,b)
except ZeroDivisionError:
    print("---division by zero!---")
else:
    print("result is", result)
finally:
    print("executing finally clause")
````

The `raise` statement is used to force a specified exception to occur

````python
raise NameError('My error')
````

If we want to assert that a logical expression is `True`, then we can use an `AssertionError`

````python
x = 1
if x == 1:
    print("Equality")
else:
    raise AssertionError("Not equal")
````

Alternatively, we can use the `assert` statement.

````python
assert x == 1
````
